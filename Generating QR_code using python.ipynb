## qrcode Module

The qrcode module in Python is used to generate QR codes from any type of data such as text, URLs, phone numbers, or JSON. 

It allows both quick QR code creation and advanced customization like size, border, color, and error correction.

This module converts the input data into a QR pattern and returns it as an image object.

```python
import qrcode

qr = qrcode.QRCode(
    version=1,
    box_size=15,
    border=5
)

qr.add_data("https://www.instagram.com/suhelsalafi/")
qr.make(fit=True)

img = qr.make_image(
    fill_color="red",
    back_color="white"
)

img.save("custom_qr.png")
```
